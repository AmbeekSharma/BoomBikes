## Step 3: Data Preparation
- In order to fit a regression line, we would need numeric values and not strings.Therefore, we need to convert them to Os and 1s

```python
#removing the columns which were created above for better visualisation
bike_file = bike_file.drop(['yr_map','workingday_map','holiday_map'], axis = 1)
bike_file.head()
```

- Creating Dummy Variables for categorical variables - season, mnth, weekday, weathersit

```python
# Dummy variables for season column
season_dummy = pd.get_dummies(bike_file['season'],dtype=int)
season_dummy.sample(10)
```

```python
# Dropping the first column from season_dummy using 'drop_first = True'
# 000 will correspond to Fall
# 100 will correspond to Spring
# 010 will correspond to Summer
# 001 will correspond to Winter

season_dummy = pd.get_dummies(bike_file['season'], prefix = 'season', drop_first = True,dtype=int)

# Adding the results to the original bike_file dataframe
bike_file = pd.concat([bike_file, season_dummy], axis = 1)

# Dropping column 'season' as we have created the dummies for it
bike_file = bike_file.drop(['season'], axis = 1)
```

```python
# Dummy variables for mnth column
mnth_dummy = pd.get_dummies(bike_file['mnth'],dtype=int)
mnth_dummy.head()
```

```python
# Dropping the first column from mnth_dummy using 'drop_first = True'
# 00000000000 will correspond to April
# 10000000000 will correspond to August
# 01000000000 will correspond to December
# 00100000000 will correspond to February
# 00010000000 will correspond to January
# 00001000000 will correspond to July
# 00000100000 will correspond to June
# 00000010000 will correspond to March
# 00000001000 will correspond to May
# 00000000100 will correspond to November
# 00000000010 will correspond to October
# 00000000001 will correspond to September

mnth_dummy = pd.get_dummies(bike_file['mnth'], prefix = 'mnth', drop_first = True,dtype=int)

# Adding the results to the original bike_file dataframe
bike_file = pd.concat([bike_file, mnth_dummy], axis = 1)

# Dropping column 'season' as we have created the dummies for it
bike_file = bike_file.drop(['mnth'], axis = 1)
```

## Step 4: Splitting the data into Training and Testing Sets

```python
#first step for regression is train test split (we can use sklearn)
from sklearn.model_selection import train_test_split
#each time we run our model, the data selected for train and test will be same by setting rantom_state to an integer
np.random.seed(0)
bike_train, bike_test = train_test_split(bike_file, train_size =0.7, test_size = 0.3, random_state = 100)
```

```python
bike_train.shape
```

```python
bike_test.shape
```

### Rescaling the features using MinMax Scaling (so that the units of coefficients obtained are all on the same scale)

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Applying scaler() to all the columns except the 'binary' and 'dummy' variables
num_vars = ['temp', 'hum', 'windspeed', 'cnt']

bike_train[num_vars] = scaler.fit_transform(bike_train[num_vars])
bike_train.head()
```

```python
bike_train.describe()
```

```python
plt.figure(figsize=(22,20))
sns.heatmap(bike_train.corr(), annot = True, cmap = "YlGnBu")
plt.show()
```

### Dividing bike_train training data set to X and Y sets for the model building

```python
# as 'cnt' is our dependent variable
y_train = bike_train.pop('cnt')
X_train = bike_train
```

```python
X_train.columns
```

## Step 5: Building a Linear model

```python
# Using the LinearRegression function from SciKit Learn for its compatibility with RFE (Recursive Feature Elimination which is a utility from sklearn)
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)

# Running RFE
#rfe = RFE(lm, 10)
rfe = RFE(estimator=lm, n_features_to_select=10)

rfe = rfe.fit(X_train, y_train)
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
# checking the columns not recommended by RFE
X_train.columns[~rfe.support_]
```

```python
# filtering out columns recommmended by RFE
rfe_fil_cols = X_train.columns[rfe.support_]
rfe_fil_cols
```

### Building model using statsmodel for detailed statistics

```python
# Creating X_train dataframe with RFE selected variables
X_train_rfe = X_train[rfe_fil_cols]
```

```python
# Adding a constant variable 
import statsmodels.api as sm
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
# Running the linear model
lm = sm.OLS(y_train, X_train_rfe).fit()
```

```python
# checking statistics using summary
print(lm.summary())
```

Looking at the summary statistics (R-squared, Adj. R-squared,  Prob (F-statistic) and P-value for variable),
the model built seems to be good but we need to make sure that the variables are not correlated. 
To check for multicollinearity, we will use VIF (Variance Inflation Factor)

```python
# Calculate the VIFs for the model using variance_inflation_factor from statsmodels
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Creating a dataframe that will contain the names of all the feature variables and their respective VIFs
X = X_train_rfe.drop(['const'], axis = 1)
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = 'VIF',  ascending = False)
vif
```

As we can notice, variable hum and temp have very high VIF values which indicates multicollinearity. 
We should drop such variables but should be dropped one by one. It can happen that after removing one such variable,
VIF of other variable can reduce.

```python
# Dropping 'hum'(VIF = 15.18) variable first  
X_train_rfe = X_train_rfe.drop(['hum'], axis = 1)
# Building the Linear model again after dropping the 'hum' variable
lm = sm.OLS(y_train, X_train_rfe).fit()
print(lm.summary())
```

```python
# Calculating the VIFs again for the model built after removing 'hum' variable
X = X.drop(['hum'], axis = 1)
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

After dropping the 'hum' variable, VIF of 'temp' variable has been reduced and is acceptable (< 5.0).

#### Also, the VIF values and p-values both are within an acceptable range. So the fitted model is good and we can make our predictions using this model.

## Step 6: Residual Analysis of train data 

To confirm the major assumption of Linear regression (to check if the error terms are normally distributed)

```python
# calculating the predicted value 
y_train_pred = lm.predict(X_train_rfe)

# to plot the histogram of the error terms
fig = plt.figure()
sns.distplot(y_train - y_train_pred, bins = 20)
fig.suptitle('Error Terms', fontsize = 20)
plt.xlabel('Errors', fontsize = 18)
plt.show()
```

This shows that errors are normally distributed around mean value 0.0

## Step 7: Making Predictions using the final model

```python
#Scaling the test data (except the dummy variables)
bike_test[num_vars] = scaler.transform(bike_test[num_vars])
bike_test.describe()
```

```python
# Dividing the test data into X_test and y_test
y_test = bike_test.pop('cnt')
X_test = bike_test
```

```python
# Now let's use our model to make predictions.

# Creating X_test_rfe dataframe by dropping variables from X_test
X_test_rfe = X_test[X.columns]

# Adding constant
X_test_rfe = sm.add_constant(X_test_rfe)

# Making predictions
y_pred = lm.predict(X_test_rfe)
```

### Model evaluation

```python
# Plotting y_test and y_pred to understand the spread.
fig = plt.figure()
plt.scatter(y_test, y_pred)
fig.suptitle('y_test vs y_pred', fontsize = 20)
plt.xlabel('y_test', fontsize = 18)
plt.ylabel('y_pred', fontsize = 16)
plt.show()
```

```python
# calculating the R-squared on the test data set
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
